### Most Common Types of Statistical Test:


#### ANOVA: 

- Compare the difference between the means of a continuous outcome, of three or more groups.


#### Paired T-Test:

- Compare the means of a continuous outcome of one group at two points of time


#### Unpaired T-Test: 

- Compare the means of a continuous outcome of two different groups.


#### Multiple linear Regression: 

- Model the relationship between two or more continuous variable and a continuous outcome.


#### Multiple logistic Regression: 

- Model the relationship between two or more continuous or categorical variables and a categorical outcome.



#### Chai-Square Test:

Comparing two categorical variables regards the number of subgroups 


#### Pearson Correlation Coefficient: 

- Describes the strength and polarity of a linear relationship between two continuous variables ( no correlation , positive correlation , negative correlation ).
