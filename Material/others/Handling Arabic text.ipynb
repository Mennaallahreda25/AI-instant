# Googletrans Overview

Googletrans is a free and unlimited python library that implemented Google Translate API. This uses the Google Translate Ajax API to make calls to such methods as detect and translate. 

Features:                  
• Fast and reliable - it uses the same servers that translate.google.com uses                  
• Auto language detection                  
• Bulk translations                       
• Customizable service URL                       
• Connection pooling (the advantage of using requests. Session)                   
• HTTP/2 support

# Simple Translation Example

```python
from googletrans import Translator
translator = Translator()
print (translator.translate( 'مرحبا').text)
print (translator.translate('مرحبا', dest='ko').text)
print (translator.translate( 'Hello' , dest='ar').text)
```

# Sentence Translation Example

```python
from googletrans import Translator
translator = Translator()

print (translator.translate("بسم الله").text)
print (translator.translate("بسم الله",dest="ko").text)
print (translator.translate("بسم الله",dest="ja").text)
print (translator.translate("بسم الله",dest="fr").text)
print (translator.translate("بسم الله",dest="zh-cn").text)
print (translator.translate("بسم الله",dest="it").text)
print (translator.translate("بسم الله",dest="ru").text)
print (translator.translate("بسم الله",dest="ur").text)
```

# Translation Object's Details

```python
from googletrans import Translator
translator = Translator()
translated = translator. translate('مرحبا منة', src='ar', dest='en')
print(" Source Language:" + translated.src)
print(" destination Language:" + translated.dest)
print(" origin:", translated.origin)
print (" Translated string:" + translated.text)
print(" Pronunciation:", translated.pronunciation)
```

# Iterate over the translation

```python
from googletrans import Translator
translator = Translator()
translations = translator.translate(['The quick brown fox','jumps over','the lazy dog'], dest='ar')
for translation in translations:
    print(translation.origin,' -> ', translation.text)
```

# Detect the Language

```python
from googletrans import Translator
translator = Translator()
print (translator.detect("بسم الله"))
print( translator.detect('이 문장은 한글로 쓰여졌습니다. '))
print(translator.detect('この文章は日本語で書かれました。'))
print (translator. detect('This sentence is written in English.'))
```

```python
from googletrans import Translator
translator = Translator()
langs = translator.detect([' مرحبا ', '한국어 ',"日本語",'English','le français'])
for lang in langs:
    print(lang.lang, lang.confidence)
```
