```python
# importing the module
from pytube import YouTube

# where to save
SAVE_PATH = "E:/" #to_do

# link of the video to be downloaded
link="https://www.youtube.com/watch?v=xWOoBJUqlbI"

try:
	# object creation using YouTube
	# which was imported in the beginning
	yt = YouTube(link)
except:
	print("Connection Error") #to handle exception

# filters out all the files with "mp4" extension
mp4files = yt.filter('mp4')

#to set the name of the file
yt.set_filename('GeeksforGeeks Video')

# get the video with the extension and
# resolution passed in the get() function
d_video = yt.get(mp4files[-1].extension,mp4files[-1].resolution)
try:
	# downloading the video
	d_video.download(SAVE_PATH)
except:
	print("Some Error!")
print('Task Completed!')
```

```python
website=requests.get("https://coreyms.com/").text
website
```

```python
soup=BeautifulSoup(website, 'lxml')
```

```python
print(soup.prettify())
```

```python
csvfile = open ('scrape5.csv', 'w', encoding='utf-8') # 'w' = write into the file
writer = csv.writer(csvfile) # import csv
writer.writerow(['headline', 'summary'])
articles = soup.find_all('article')

for i in articles :
    headline = i.h2.a.text
    summary = i.find('div', class_ ='entry-content').p.text
    writer.writerow([headline, summary])    
        
csvfile.close()
```

```python
website=requests.get("https://www.youtube.com/feed/channels").text
website 
```

```python
soup=BeautifulSoup(website, 'lxml')
print(soup.prettify())
```

```python
csvfile = open ('scrape.csv', 'w', encoding='utf-8') # 'w' = write into the file
writer = csv.writer(csvfile) # import csv
writer.writerow(['headline', 'link'])
articles = soup.find_all('article')

for i in articles :
    headline = i.h2.a.text
    link = "None"
    
    try:
        vid = i.find('iframe', class_ = "youtube-player")['src']
        vid_id = vid.split('/')
        video = vid_id[4].split('?')
        c = video[0]
        link = f'https://youtube.com/watch?v={c}'
    except Exception as e:
        link = "No Link"
        
    writer.writerow([headline, link ])    
        
csvfile.close()            
```
