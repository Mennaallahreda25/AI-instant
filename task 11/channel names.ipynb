```python
import requests
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
```

```python
channels=[]
summary=[]
```

```python
with open('channels.htm', encoding='utf-8') as html_file:
    soup = BeautifulSoup(html_file, 'lxml') #parser # parsing
soup.content
```

```python
print(soup.prettify())  # to prettify the generated HTML
```

```python
channels=soup.find("ytd-app")
channels
```

```python

name_channels=channels.find_all("yt-formatted-string",{"class":"style-scope ytd-channel-name"})
print(name_channels)

summary=channels.find_all("yt-formatted-string", class_="style-scope ytd-channel-renderer")
summary
```

```python
for i in range(len(channels)):
    channels.append(channels[i].yt-formatted-string.text)
    summary.append(summary[i].yt-formatted-string.text)
file_list=[channels,summary]
exported=zip_longest(*file_list)
with open("/users/hp/mychannels.csv",'w') as myfile:
    wr=csv.writer(myfile)
    wr.writerow(["channels","summary"])
    wr.writerows([exported])
```
